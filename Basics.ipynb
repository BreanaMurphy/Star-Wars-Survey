```python
import pandas as pd
star_wars=pd.read_csv("star_wars.csv",encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars=star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head()
```

```python
yes_no={"Yes":True,"No":False}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]=star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]=star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

star_wars.head()
```

```python
import numpy as np
movie_map={"Star Wars: Episode I  The Phantom Menace":True, np.NaN: False,"Star Wars: Episode II  Attack of the Clones":True,"Star Wars: Episode III  Revenge of the Sith":True,"Star Wars: Episode IV  A New Hope":True,"Star Wars: Episode V The Empire Strikes Back":True,"Star Wars: Episode VI Return of the Jedi":True}
for col in star_wars.columns[3:9]:
    star_wars[col]=star_wars[col].map(movie_map)

star_wars=star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1","Unnamed: 4":"seen_2","Unnamed: 5":"seen_3","Unnamed: 6":"seen_4","Unnamed: 7":"seen_5","Unnamed: 8":"seen_6"})
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
star_wars=star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1","Unnamed: 10":"ranking_2","Unnamed: 11":"ranking_3","Unnamed: 12":"ranking_4","Unnamed: 13":"ranking_5","Unnamed: 14":"ranking_6"})
star_wars[star_wars.columns[9:15]].head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

means=np.mean(star_wars[star_wars.columns[9:15]])
plt.bar(range(6),means)
```

On Average individuals liked movie 5 best (lower rankings are better) and they like movie 3 the least. 

```python
seen=star_wars[star_wars.columns[3:9]].sum()
plt.bar(range(6),seen)
```

The above shows that the older movies are more popular than the newer movies.

```python
males=star_wars[star_wars["Gender"]=="Male"]
females=star_wars[star_wars["Gender"]=="Female"]
fig=plt.figure(figsize=(10,8))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.set_title("Males that have seen each movie")
ax1.bar(range(6),males[males.columns[3:9]].sum())
ax2.set_title("Females that have seen each movie")
ax2.bar(range(6),females[females.columns[3:9]].sum())
ax3.set_title("Average Male Ranking of Star Wars Movies")
ax3.bar(range(6),males[males.columns[9:15]].mean())
ax4.set_title("Average Femal Ranking of Star Wars Movies")
ax4.bar(range(6),females[females.columns[9:15]].mean())
plt.show()
```

The distributions for Male and Females for are very similiar, but there does appear to be more men that have seen each movie in our survey (this could be because more men responded to the survey).
Women tended to rate movies 1 and 2 lower than men and they rates movie 4 higher than men.
